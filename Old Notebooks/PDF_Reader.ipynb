```
import requests 
from io import BytesIO
import re 
import PyPDF2
import os
```

```
# api-endpoint 
URL = "https://www.osti.gov/api/v1/records"
  
# defining a params dict for the parameters to be sent to the API 
# PARAMS = {'research_org':"Argonne National Laboratory ANL","rows":3200} 
# PARAMS = {'research_org':"Advanced Light Source ALS","rows":3200} 
PARAMS = {'research_org':"Lawrence Berkeley National Lab","rows":3200}

# sending get request and saving the response as response object 
r = requests.get(url = URL, params = PARAMS) 
  
# extracting data in json format 
data = r.json() 
```

```
links = []

for obj in data:
  if(len(obj["links"]) > 1):
    links.append(str(obj["links"][1]["href"])+"?directFulltextAccess=LBNL")
```

```
from io import StringIO

from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfparser import PDFParser

publications = {}
# dir = "/content/drive/MyDrive/LBNL/osti_new/"
dir = "/content/drive/MyDrive/LBNL/osti_anl/"
files = os.listdir(dir)

for file in files:
  file_name = dir + file
  print(file_name)
  output_string = StringIO()
  with open(file_name, 'rb') as in_file:
      parser = PDFParser(in_file)
      doc = PDFDocument(parser)
      rsrcmgr = PDFResourceManager()
      device = TextConverter(rsrcmgr, output_string, laparams=LAParams())
      interpreter = PDFPageInterpreter(rsrcmgr, device)
      for page in PDFPage.create_pages(doc):
          interpreter.process_page(page)

  publications[file] = output_string.getvalue()
```

```
cleaned_publications = {}

for key,publication in publications.items():
  publication = publication.lower()
  begin = publication.find("abstract")
  end  = publication.find("acknowledgments")
  if(end == -1):
    end  = publication.find("acknowledgements")
  if(end == -1):
    end  = publication.find("acknowledgment")
  if(end == -1):
    end  = publication.find("acknowledgement")

  ref  = publication.find("references")
  if(ref != -1 and ref < end):
    end = ref
    
  if(end == -1):
    end = len(publication)-1

  print(key, end)
  content = publication[begin:end]
  cleaned_publications[key] = content
```
